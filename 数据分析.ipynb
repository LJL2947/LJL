```python
def IQR_function(df):
    IQR = []
    upper = []
    lower = []
    max_i = []
    min_i = []
    for i in df.columns:
        Q1 = np.percentile(df[i],25,interpolation='midpoint')
        Q3 = np.percentile(df[i],75,interpolation='midpoint')
        IQR.append(Q3-Q1)
        upper.append(Q3+1.5*(Q3-Q1))
        lower.append(Q1-1.5*(Q3-Q1))
        max_i.append(df[i].max())
        min_i.append(df[i].min())
    iqr = pd.DataFrame([IQR,upper,lower,max_i,min_i]).T
    iqr.columns = ['IQR','upper','lower','max','min']
    iqr.index = df.columns   
    return iqr
```

```python
def out_line_function(df, iqr):
    out_line = []
    for i in df.columns:
        upper_out = df[i]>iqr['upper'][i]
        lower_out = df[i]<iqr['lower'][i]
        out_line.extend(df[upper_out].index) 
        out_line.extend(df[lower_out].index)
    return out_line
```

```python
iqr = IQR_function(train_data)
iqr
```

```python
out_line = out_line_function(train_data, iqr)
```

```python
len(np.unique(out_line))
```

```python
train_data = train_data.drop(index=out_line,axis=0)
train_data.index = range(len(train_data))
```
